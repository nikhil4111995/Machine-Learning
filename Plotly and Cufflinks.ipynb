```python
#spread type visualisation is maily for stock data
#comparison between two stocks
```

```python
#we'll get a plot and then a subplot I'll get my line plot which shows them against each other and
#then will get my spread plot which shows the spread of them against each other.
#So for instance if we go ahead and zoom in here on the 20 we can zoom in here.Let's see where it corresponds to 20.
#Well notice the spread is negative .38 between them.And here we can verify that as a result.
#So negative .38 minus point 1.We'll take a 2.56 about there.
#So you can go ahead and compare these results and this is essentially really useful for stock data to
#see the spread between two stocks.
```

```python
df[['A','B']].iplot(kind='spread')
```

```python
#a bubble plot is very similar to a scatterplot except that it will change size of the points based off 
#of another variable.
```

```python
df.iplot(kind ='bubble',x='A',y='B',size='C')
#we have a scatter plot except the size of the points is based off in other column value
#and you see these kind of plots a lot for things like world GDP in comparison to population and maybe
#happiness factor etc.So these are quite common plots especially for things like United Nations reports.
```

```python
#scatter matrix
#we'll get to the scatter matrix the matrix plot is essentially very similar to seaborne pair plot.
#It just creates a scatter matrix of all the columns it can. You need to make sure all your columns
#are numerical in order for this to work.
```
