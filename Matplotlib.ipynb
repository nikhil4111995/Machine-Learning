```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
#There are two ways to create plots. One is functional method and other one in object oriented method
```

```python
#Functional method
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
#Mulitplots on the same canvas
```

```python
#subplot(number of rows,columns,plot number you are referring to)
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#Object oriented API method
```

```python
#we are going to instantiate figure objects and then call methods or attributes from that object
#firstly we create an empty figure and then we add axis to it.
```

```python
fig = plt.figure()
#add_axes([left,bottom,width,height]) takes 4 arguments. Basically the percent of the blank canvas you want to go 
#ahead and take
#make sure the numbers in list are between 0 and 1 because we are taking as percentage of canvas area 
axes = fig.add_axes([0.1,0.1,0.8,0.8])
#plotting on the axes
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
#Putting two set on one canvas(subplot)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
#[0.2,0.5,0.4,0.3] means 20% in from the left, 50% up from the bottom,40% of the canvas size and width 
#30% of the canvas size and height 
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
#plot on these axes
axes1.plot(x,y)
axes1.set_title('Large plot')
axes2.plot(y,x,'r')
axes2.set_title('Small plot')
```

```python
#check list objects refering to columns and rows
```

```python
#Creating subplots using OOP
```

```python
axes
```

```python
#Important: plt.subplot adding axes automatically based on rows and columns whereas in plt.figure we are adding axes manually
#fig,axes tuple unpacking
fig,axes = plt.subplots(nrows=1,ncols=2)#means 1 row 2 columns displayed as 2 plots
#we can iterate through axes objects since it is a list as just mentioned above
#you can plot by indexing or iterating using for loop
#use axes object to plot on it
axes[0].plot(x,y)
axes[0].set_title('Plot 1')
axes[1].plot(y,x)
axes[1].set_title('Plot 2')

```

```python
#legend: it differentiate the lines created on the same plot, to avoid the confusion
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2)
ax.plot(x,x**3)
#Without refering to code it will be confusing to find out what each plot represents 
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
#to make legend funciton to work, we need to add label argument 
ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='Y Cubed')
#sometimes legend overlaps the actual plot, to avoid this, specify loc 0 to 10, 
#0 represent best location, so it is better to use that
#if you cant find the match you can manually type the location of the legend. For ex: ax.legend(loc=(0.1,0.1))

ax.legend(loc=0)
```

```python
#Setting colors with matplotlib
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
#alpha argument for more transparency on the line
#instead of 'linewidth' you write 'lw' as well
#'linestyle'(ls) is for dash line or dotted line,check for jupter notebook for more options
#marker is going to be used, when we have few data points. It marks out point at x versus y
ax.plot(x,y,color='purple',linewidth=1,alpha = 1,linestyle='-',marker='o',markersize=10,markerfacecolor='yellow'
       ,markeredgewidth = '3',markeredgecolor='green')#you can also put RGB hex code. Ex: #FF8C00
#check jupyter notebook 
```

```python
#control over axis appearacne and plot range
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',lw=2,ls='--')
#showing plot betweeen 0 and 1 on the x axis
ax.set_xlim([0,1])#upperbounnd ,lower bound
ax.set_ylim([0,2])
```
